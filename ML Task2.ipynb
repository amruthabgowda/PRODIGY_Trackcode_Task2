```python
# Set the environment variable to avoid memory leak issues
import os
os.environ['OMP_NUM_THREADS'] = '1'

# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

# Load the customer purchase history data (replace with your actual dataset)
data = {
    'CustomerID': range(1, 21),  # 20 customers
    'TotalPurchaseAmount': [100, 200, 300, 400, 500, 150, 250, 350, 450, 550,
                             120, 220, 320, 420, 520, 170, 270, 370, 470, 570],
    'PurchaseFrequency': [5, 10, 3, 8, 12, 6, 9, 4, 7, 11,
                          5, 10, 3, 8, 12, 6, 9, 4, 7, 11]
}
df = pd.DataFrame(data)

# Preprocess the data
X = df[['TotalPurchaseAmount', 'PurchaseFrequency']]
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Apply K-means clustering
kmeans = KMeans(n_clusters=4, random_state=42)  # Adjust the number of clusters (n_clusters) as needed
df['Cluster'] = kmeans.fit_predict(X_scaled)

# Visualize the clusters
plt.figure(figsize=(10, 8))
plt.scatter(df['TotalPurchaseAmount'], df['PurchaseFrequency'], c=df['Cluster'], cmap='viridis', marker='o', s=100)
plt.title('Customer Segmentation Based on Purchase History')
plt.xlabel('Total Purchase Amount')
plt.ylabel('Purchase Frequency')
plt.colorbar(label='Cluster')
plt.show()

# Analyze the results
print(df[['CustomerID', 'TotalPurchaseAmount', 'PurchaseFrequency', 'Cluster']])
```
